```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# seed 값 설정

seed=0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
# mnist 데이터셋 불러오기

(x_train,y_class_train),(x_test,y_class_test)=tf.keras.datasets.mnist.load_data()

print('학습셋 이미지수: {} 개' .format(x_train.shape[0])) # minist 손글씨 학습셋 이미지셋 은 (60000,28,28) 로 구성되어 있다
print('테스트셋 이미지 수: {} 개' .format(x_test.shape[0]))

```

```python
# 차원 변환 과정

x_train=x_train.reshape(x_train.shape[0],784).astype(float) /255 # 이제 주어진 가로 28, 세로 28개의 2차원 배열을 784개의 1차원 배열로 바꿔 줘야 한다. # 흰색이 0이고 글쓰기 1~255까지이기 때문에 0부터 1 사이를 만들기 위해서 255으로 나눠 준다. 컨볼루션층이나 맥스풀링은 주어진 이미지를 2차원 배열인채로 다룬다는 점이다. 활성화 함수는 1차원 배열로 다룬다

print(x_train.shape)
```

```python
x_test=x_test.reshape(x_test.shape[0],784).astype(float) / 255 # 이제 주어진 가로 28, 세로 28개의 2차원 배열을 784개의 1차원 배열로 바꿔 줘야 한다. # 흰색이 0이고 글쓰기 1~255까지이기 때문에 0부터 1 사이를 만들기 위해서 255으로 나눠 준다

print(x_test.shape)
```

```python
y_train= pd.get_dummies(y_class_train) # 0부터 9까지 숫자로 되어 있는 것도 onehot코딩을 해 줘야 한다
y_test= pd.get_dummies(y_class_test) # 0부터 9까지 숫자로 되어 있는 것도 onehot코딩을 해 줘야 한다
```

```python
x=tf.keras.layers.Input(shape=[784])
h=tf.keras.layers.Dense(512,activation='relu')(x)
h=tf.keras.layers.Dense(10,activation='relu')(h)
h=tf.keras.layers.Dense(10,activation='relu')(h)
y=tf.keras.layers.Dense(10, activation='softmax')(h)

```

```python
model=tf.keras.models.Model(x,y)
model.compile(loss='categorical_crossentropy',optimizer='adam', metrics='accuracy')
```

```python
from keras.callbacks import EarlyStopping
Early_Stopping_callbacks=EarlyStopping(monitor='accuracy', patience=50) # 모니터할 값과 테스트 오차가 좋아지지 않아도 몇번까지 기다릴지 결정한다

```

```python
model.fit(x_train,y_train,epochs=5000, batch_size=100, callbacks=[Early_Stopping_callbacks])
```
