```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mydata = 'C:/Users/doimo/pythonDataWorkspace/080228-master/080228-master/deeplearning/dataset/pima-indians-diabetes.csv'
df = pd.read_csv(mydata, names=["pregnant", "plasma", "pressure", "thickness", "insulin", "BMI", "pedigree", "age", "class"])

# print(df.shape)
# print(df.columns)
print(df.head(5))
print(df.info())
```

```python
colormap=plt.cm.gist_heat #   .
plt.figure(figsize=(20,20)) #  .

sns.heatmap(df.corr(),linewidths=0.1,vmax=0.5, cmap=colormap, linecolor='white', annot=True) #   . vmax  0.5  0.5      . cmap   , linecolor , linewidth  
plt.show()

#        class    . plasma()        
```

```python
import glob
import tensorflow as tf
import numpy as np
```

```python

#        .
np.random.seed(3)
tf.random.set_seed(3)


#   .
dataset = np.loadtxt(mydata, delimiter=",")
independent = dataset[:,0:8]
sub = dataset[:,8]


#  

x = tf.keras.layers.Input(shape=[8])
y = tf.keras.layers.Dense(1)(x)
model= tf.keras.models.Model(x,y)
model.compile(loss='mse', metrics='accuracy')

#  

model.fit(independent,sub, epochs=10000)


```
