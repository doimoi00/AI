```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mydata = 'C:/Users/doimo/pythonDataWorkspace/080228-master/080228-master/deeplearning/dataset/pima-indians-diabetes.csv'
df = pd.read_csv(mydata, names=["pregnant", "plasma", "pressure", "thickness", "insulin", "BMI", "pedigree", "age", "class"])

# print(df.shape)
# print(df.columns)
print(df.head(5))
print(df.info())
```

```python
colormap=plt.cm.gist_heat #그래프의 색상 구성을 정합니다.
plt.figure(figsize=(20,20)) #그래프의 크기를 정합니다.

sns.heatmap(df.corr(),linewidths=0.1,vmax=0.5, cmap=colormap, linecolor='white', annot=True) # 그래프의 속성을 결정합니다. vmax의 값을 0.5로 지정해 0.5에 가까울 수록 밝은 색으로 표시되게 합니다. cmap 은 색깔 테마, linecolor 테두리선, linewidth 테두리 굵기
plt.show()

# 아래 표에 다른 것은 별로 안 중요하지만 class가 결과이기 때문에 가장 중요하다. plasma(공복혈당수치)가 가장 상관 관계까 높다는 것을 알 수 있다
```

```python
import glob
import tensorflow as tf
import numpy as np
```

```python

# 실행할 때마다 같은 결과를 출력하기 위해 설정하는 부분입니다.
np.random.seed(3)
tf.random.set_seed(3)


# 데이터를 불러 옵니다.
dataset = np.loadtxt(mydata, delimiter=",")
independent = dataset[:,0:8]
sub = dataset[:,8]


# 모델을 만듭니다

x = tf.keras.layers.Input(shape=[8])
y = tf.keras.layers.Dense(1)(x)
model= tf.keras.models.Model(x,y)
model.compile(loss='mse', metrics='accuracy')

# 모델을 학습합니다

model.fit(independent,sub, epochs=10000)


```
