```python
import numpy as np
import tensorflow as tf
```

```python
# seed 값을 설정한다
np.random.seed(3)
tf.random.set_seed(3)
```

```python
#  데이터를 입력한다

import pandas as pd

df=pd.read_csv('C:/Users/doimo/pythonDataWorkspace/080228-master/080228-master/deeplearning/dataset/sonar.csv', header=None) #  read_csv 의 인자중 header 를 None 으로 주면 판다스는 인덱스를 자동으로 만들어낸다.
print(df.info())
```

```python
print(df.head())
```

```python

dataset=df.values # [중혁] 다른 곳에서는 values를 안 해도 되었는데 여기서는 왜 꼭 해 줘야 하는지 모르겠다
ind=dataset[:,0:60].astype(float) #[중혁] tensorflow를 돌릴 때는 float은 float이라고 지정해 줘야 하는 거 같다
sub=dataset[:,60]

sub=pd.get_dummies(sub) # [중혁] onehot을 해서 y_obj가 한개가 아니라 문자  개수만큼 변한다 (208, 2)가 된다. 그렇기 때문에 y가 2가 되어야 한다

x=tf.keras.layers.Input(shape=[60])
h=tf.keras.layers.Dense(8,activation='swish')(x)
h=tf.keras.layers.Dense(8,activation='swish')(h)
h=tf.keras.layers.Dense(8,activation='swish')(h)
y=tf.keras.layers.Dense(2,activation='sigmoid')(h) #[중혁] get_dummies를 하면 그 개수만큼 나와야 하는거 같음
model=tf.keras.models.Model(x,y)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')

model.fit(ind,sub,epochs=1000)


```
