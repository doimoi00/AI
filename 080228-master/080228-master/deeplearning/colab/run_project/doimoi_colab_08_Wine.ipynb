```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)

#데이터 입력

df_pre=pd.read_csv('C:/Users/doimo/pythonDataWorkspace/080228-master/080228-master/deeplearning/dataset/wine.csv', header=None) # 랜덤하게 섞어 줄 데이터를 먼저 df_pre에 넣어 놓는다
df=df_pre.sample(frac=1)
dataset=df.values

print(dataset.shape)
```

```python
ind=dataset[:,0:12]
sub=dataset[:,12]

print(ind.shape)
print(sub.shape)
```

```python
x=tf.keras.layers.Input(shape=[12])
h=tf.keras.layers.Dense(30,activation='relu')(x)
h=tf.keras.layers.Dense(12,activation='relu')(h)
h=tf.keras.layers.Dense(8,activation='relu')(h)
y=tf.keras.layers.Dense(1,activation='sigmoid')(h)
```

```python
model=tf.keras.models.Model(x,y)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')


from keras.callbacks import EarlyStopping
Early_Stopping_callbacks=EarlyStopping(monitor='val_loss', patience=100) # 모니터할 값과 테스트 오차가 좋아지지 않아도 몇번까지 기다릴지 결정한다

model.fit(ind,sub, validation_split=0.2, epochs=2000, batch_size=200, callbacks=[Early_Stopping_callbacks])
# 학습셋 오차가  loss 이다. 테스트 오차는 val_loss 이다. 학습 정확느는 acc, 테스트셋 정확도는 val_acc, 학습셋 오차는 loss
```
